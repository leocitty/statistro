```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
N = 100000
porte = 10
M = 6
gioccate = np.zeros((1, 4))
markall = 0
jeffal = 0
frankall = 0

for i in range(1, N+1):
    door = [0]*porte
    lambo = rd.randint(0,(porte-1))
    door[lambo] = 1

    questa = rd.randint(0,(porte-1))
    jeff = questa
    mark = questa
    for j in range(0, M):
        capra = rd.randint(0,(porte-1))
        while (capra==questa or capra==lambo or door[capra]==2):
            capra = rd.randint(0,(porte-1))
        door[capra] = 2

    while (jeff==questa or door[jeff]==2):
        jeff = rd.randint(0,(porte-1))

    frank = rd.randint(0,(porte-1))
    while (door[frank]==2):
        frank = rd.randint(0,(porte-1))
    
    markall = markall + door[mark]
    jeffal = jeffal + door[jeff]
    frankall = frankall + door[frank]
    if i%10 == 0:
        gioccate = np.append(gioccate, [[i, markall, jeffal, frankall]], axis=0)
        
```

```python
print(gioccate[-1, :])
gioccate = gioccate[1:, :]
plt.errorbar(gioccate[:,0], gioccate[:,1]/gioccate[:,0], label='mark')
plt.errorbar(gioccate[:,0], gioccate[:,2]/gioccate[:,0], label='jeff')
plt.errorbar(gioccate[:,0], gioccate[:,3]/gioccate[:,0], label='frank')
plt.legend()
```
