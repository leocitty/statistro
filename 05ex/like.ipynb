```python
import astropy
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.stats import norm
from numpy import random as rd
```

```python
def likehood(x, med, sig):
    return (sig*np.sqrt(2*np.pi))*np.exp(-((x-med)**2)/(2*sig**2))
```

```python
#PART 1

N = 100
quaz = rd.normal(loc=1, scale=0.2, size=N)
quaz = np.sort(quaz)
hoodlik = np.zeros(N)
for i in range(0,N):
    hoodlik[i] = likehood(quaz[i], 1, 0.2)

totlikhood = np.prod(hoodlik)
plt.scatter(quaz, hoodlik)
```

```python
print('x MLE:', np.mean(quaz))
print('x with max L:', quaz[np.where(hoodlik==max(hoodlik))[0][0]])
```

```python
#PART 2

loghood = np.log(hoodlik)
#plt.scatter(quaz, loghood)

#difflog = (np.diff(loghood, 2)/np.diff(quaz, 2)**2)
difflog = np.gradient(loghood, 2)
difflog
plt.scatter(quaz, difflog)
```
